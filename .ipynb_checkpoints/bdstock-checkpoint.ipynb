```python
# Importing the dataset
dataset = pd.read_csv('JANATA.csv')
```

```python
# if your dataset contains missing value, remove those missing values
dataset.dropna(inplace=True)
```

```python
dataset.set_index('date', inplace=True)
```

```python
StockPrice = dataset['closing_price']
dataset.tail()
```

# Sliding Window

```python
historical_days = []
this_day = []
step = 1
max_len = 5 ### lagged variable/ Sliding Window
for idx in range(0, len(StockPrice) - max_len, step):
    historical_days.append(StockPrice[idx: idx + max_len])
    this_day.append(StockPrice[idx + max_len])

x = np.array(historical_days)
y = np.array(this_day)
```

```python
x.shape
```

```python
plt.plot(dataset.closing_price)
```

# Train/Test Split

```python
split = int(len(dataset) * 0.8)  
X_train = x[:split]
X_test = x[split:]
y_train = y[:split]
y_test = y[split:]
print("Training/Testing Size")
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
#splitting data into training &testing

data_training =pd.DataFrame(dataset['closing_price'][0:int(len(dataset)*0.70)])
data_testing =pd.DataFrame(dataset['closing_price'][int(len(dataset)*0.70):int(len(dataset))])


plt.plot(data_training,label="training data")
plt.plot(data_testing, label="testing data")
plt.title("Train/Test Split")
plt.legend()
plt.show()
```

# Performance Measure

```python
def root_mean_squared_error(y_true, y_pred):
        return np.sqrt(sum((y_pred-y_true)**2)/len(y_true)) 
```

```python
def mean_absoulate_percentage_error(y_true, y_pred):
        return np.mean(np.abs((y_pred-y_true) / y_true))
```

```python
from sklearn.metrics import make_scorer
def rmse(actual, predict):
    predict = np.array(predict)
    actual = np.array(actual)
    distance = predict - actual
    square_distance = distance ** 2
    mean_square_distance = square_distance.mean()
    score = np.sqrt(mean_square_distance)
    return score
rmse_score = make_scorer(rmse, greater_is_better=False)
```

```python
from sklearn.metrics import make_scorer
def mape(actual, predict):
    score = mean_absoulate_percentage_error(actual, predict)
    return score
mape_score = make_scorer(mape, greater_is_better=False)
```

# Decision Tree (Without Hyperparameter Tuning)

```python
modelDT_0 = DecisionTreeRegressor(random_state=1000)
modelDT_0.fit(X_train, y_train)
y_pred_DT_0 = modelDT_0.predict(X_test)
print('RMSE: %.4f'% np.sqrt(sum((y_pred_DT_0-y_test)**2)/len(y_test)))
print('MAPE: %.4f'% np.mean(np.abs((y_test - y_pred_DT_0) / y_test)))
```

```python
sns.set(style='whitegrid', palette='muted', font_scale=1.5)
rcParams['figure.figsize'] = 16, 10
rcParams['figure.figsize'] = 16, 10

plt.plot(y_test, marker='.', label="Actual Price")
plt.plot(y_pred_DT_0, 'r', label="Decision Tree prediction")
plt.ylabel('Stock Price in JANATA')
plt.xlabel('Time Step')
plt.title('Decision Tree forecasting for Bangladesh Stock Price ')
plt.legend()
plt.show();
```

```python
filename = 'decision_tree'
pickle.dump(modelDT_0 ,open(filename,'wb'))
```

# Decision Tree

```python
tscv = TimeSeriesSplit(9)
modelDT = DecisionTreeRegressor(random_state=1000)
param_grid_DT = {'splitter': ['best','random'],  
                'max_features': ['auto','sqrt','log2'], 
                'min_samples_split': [2, 4, 6, 8],  
                'min_samples_leaf': [1, 2, 3, 5]
                }

grid_searchDT = GridSearchCV(estimator = modelDT, param_grid = param_grid_DT, cv = tscv,scoring=mape_score, n_jobs=-1, verbose = 1)

grid_searchDT.fit(X_train, y_train)
best_paramsDT = grid_searchDT.best_params_
#pprint(grid_searchET.cv_results_)
print(best_paramsDT)
print(grid_searchDT.best_score_)
print(grid_searchDT.best_estimator_)
TunedDT = grid_searchDT.best_estimator_
```

```python
y_pred_DT = TunedDT.predict(X_test)
print('RMSE: %.4f'% np.sqrt(sum((y_pred_DT-y_test)**2)/len(y_test)))
print('MAPE: %.4f'% np.mean(np.abs((y_test - y_pred_DT) / y_test)))
```

```python
sns.set(style='whitegrid', palette='muted', font_scale=1.5)
rcParams['figure.figsize'] = 16, 10
rcParams['figure.figsize'] = 16, 10

plt.plot(y_test, marker='.', label="Actual Price")
plt.plot(y_pred_DT, 'r', label="Decision Tree prediction")
plt.ylabel('Stock Price in JANATA')
plt.xlabel('Time Step')
plt.title('Decision Tree forecasting for Bangladesh Stock Price ')
plt.legend()
plt.show();
```
